```python
from pandas.api.types import is_string_dtype,is_numeric_dtype

def str_to_cat(my_df):
    for p,q in my_df.items():
        print(my_df.items)
        if is_string_dtype(q):
            my_df[p] = q.astype('category').cat.as_ordered()
    return my_df


my_df = str_to_cat(dataset)

def mydf_to_nums(my_df, feature, null_status):
    if not is_numeric_dtype(feature):
        my_df[null_status] = feature.cat.codes
        
for p,q in dataset.items(): 
        mydf_to_nums(dataset, q, p)
```

```python
# X_cat -> catagorical, X_con -> continuous
df_cat = dataset[['CASE_STATUS', 'EMPLOYER_NAME', 'SOC_NAME', 'JOB_TITLE', 'WORKSITE','Y',2012.0,2013.0,2014.0,2015.0,2016.0]]
df_con = dataset.drop(df_cat,axis = 1)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(df_con)
df_con_sc = pd.DataFrame(scaler.transform(df_con))
df_con_sc.columns = ['PREVAILING_WAGE']
```

```python
df_cat.reset_index(inplace = True,drop = False)
df_cat.drop("index",inplace = True,axis = 1)
df_list = [df_cat,df_con_sc]
df_full = pd.concat(df_list,axis = 1)
df_full.head(5)
```

```python
df_full.head(20)
```

```python
X = df_full.iloc[:, 1:]
Y = df_full.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(X,Y,train_size = 0.7,random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
"""from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True],
    'max_depth': [3,5],
    'min_samples_leaf': [3,5],
    'n_estimators': [40,60,80,100],
    'criterion': ["gini","entropy"]
}
"""

from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True],
    'max_depth': [5],
    'min_samples_leaf': [3],
    'n_estimators': [40],
    'criterion': ["gini"]
}

grid_search_rf = GridSearchCV(RandomForestClassifier(), param_grid, cv=3, n_jobs=-1)
grid_search_rf.fit(x_train, y_train)
print(grid_search_rf.best_score_)
print(grid_search_rf.best_params_)
print(grid_search_rf.best_estimator_)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

y_pred = grid_search_rf.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
```

Here we got the accuracy as 96% but the model is useless because the data is imbalanced as we can see that from the above bar graph...


And this is the reason why we got lot of False Negative values


So in order to balance the we can either use over sampling or under sampling...but over sampling is more efficient than the under sampling...

```python
df_full.rename(columns = {2012.0:'2012',2013.0:'2013',2014.0:'2014',2015.0:'2015',2016.0:'2016'}, inplace = True)
```

```python
X = df_full.iloc[:, 1:]
Y = df_full.iloc[:, 0]
```

```python
from imblearn.over_sampling import RandomOverSampler as ra
```

```python
over_sampling = ra()
x_train_res,y_train_res = over_sampling.fit_sample(X,Y)
```

```python
x_train_res.shape,y_train_res.shape
```

```python
x_train,x_test,y_train,y_test = tts(x_train_res,y_train_res,train_size = 0.7,random_state=42)
```

```python

from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True],
    'max_depth': [5],
    'min_samples_leaf': [3],
    'n_estimators': [40],
    'criterion': ["gini"]
}

grid_search_rf = GridSearchCV(RandomForestClassifier(), param_grid, cv=3, n_jobs=-1)
grid_search_rf.fit(x_train, y_train)
print(grid_search_rf.best_score_)
print(grid_search_rf.best_params_)
print(grid_search_rf.best_estimator_)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred = grid_search_rf.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
